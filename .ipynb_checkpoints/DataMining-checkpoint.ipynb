## Methods

To investigate how brewing trend have changed over time, I first had to find relevant data sources. I was able to find data on the number of breweries broken down into sizes based on the number of barrels they were producing each year from the Alcohol and Tobacco Tax and Trade Bureau. The same site also had data listed seperately on the materials being used to produce in each month of each year from the year 2008-2014. This included data on hops, malted barley and various other materials being used. 

The other data source was the United States Department of Agriculture. From their website I was able to obtain PDFs from each each year containing data on hop production in the US from 2008-2019. However, the data file from 2013 was missing. 

To download the data I used the requests library to save each and every PDF locally. This happens in a function called save_pdfs_locally. The function gets called in store_all_files_locally which also contains a list of the URLs.

```python
FileAcquire.store_all_files_locally()
```

Then, in order to convert the files into a format that I could work with I used tabula-py which is a python wrapper of tabula-java. Tabula has the capability to extract tables from PDFs. I wrote this into a function called pdf_to_dataframe. An arguments in the function lets you decide if you want to look for gridlines or whitespace between numbers. Another allows you to choose an area on the PDF you want tabula to focus on and ignore all other parts of the page. The tables were then all stored as csv files. 

```python
pdf_names = ["Materials2008","Materials2009","Materials2010","Materials2011","Materials2012","Materials2013","Materials2014"]
FileAcquire.pdf_to_dataframe(pdf_names,"PDFData","ExtractedCSV",(385.178, 34.043, 595.553, 581.783))
```

```python
pdf_names = ["NumBrewers2008","NumBrewers2009","NumBrewers2010","NumBrewers2011","NumBrewers2012","NumBrewers2013","NumBrewers2014"]
FileAcquire.pdf_to_dataframe(pdf_names,"PDFData","ExtractedCSV")
```

```python
pdf_names = ["Hops2008","Hops2009","Hops2010","Hops2011","Hops2012","Hops2014","Hops2015","Hops2016","Hops2017","Hops2018","Hops2019"]
FileAcquire.pdf_to_dataframe(pdf_names,"PDFData","ExtractedCSV",stream = True)
FileAcquire.clean_hop_csv()
```
